## 자료형

- 변수에 이름을 붙여서 메모리에 데이터를 할당
- 메모리에 저장된 데이터는 다양한 종류가 있음
  - 예) 사람의 나이 : 숫자,
  주소 : 문자


- 모든 프로그래밍들에서 데이터 유형(data type)을 자료형으로 처리


  - numbers
  - string
  - list
  - tuple
  - dictionary


- 자료형이 필요한 이유
  - 적절한 그릇을 사용해야 부족함이나 낭비가 없음
  

- Number

 - number 타입은 숫자의 값을 저장
 - int (부호 있는 정수)
 - float (실수)


- Strings

     - 문자열은 인용부호 사이에 있는 연속된 집합으로 식별 : " ", ' ',
     - '+' 기호 : 문자를 더함
     - '*' 기호 : 문자를 반복


- List

    - 복합 데이터 유형에 가장 유용
    - 리스트 내의 항목은 쉼표(,)로 구별, [] 안에 포함

    - 리스트 인덱스는 0 부터 시작
    - '+' : 연결 연산자
    - '*' : 반복 연산자


- Tuple

   - 리스트와 유사
   - 튜플 내의 항목은 모두 쉼표(,)로 구분,
   - ( ) 안에 포함

   - 튜플의 인덱스는 0부터 시작
   - "+" 연결 연산자
   - "*" 반복 연산자
   - 리스트와의 차이점 : ** 읽기 전용**


- Dictionary

   - 쌍으로 만든 테이블 형식의 일종
   - key - value  구성
   - key : 문자열, 숫자를 많이 사용
   - value : 모든 유형의 임의의 파이썬 객체

리스트 : [서울 , 노원구]
튜플 : (서울, 노원구)
딕셔너리 : {"apple" : "사과"}

# 함수

   1. 함수란 무엇인가

  - 반복적으로 사용되는 부분을 묶어서, **재사용** 가능하도록 만들어 주는 것
  - 함수에는 input, output
  - 정해진 규칙에 따라 input -> output

```python
# 예시

#비효율적 3번 반복

x, y, z = 1, 2, 3
(x + y) * z
```

```python
x, y, z = 10, 20, 30

(x + y) * z
```

```python
x, y, z = 2, 3, 4

(x + y) * z
```

- 이 예시를 함수로 변경하여 적용해보자

```python
def func(x, y, z) :
  return (x + y) * z
```

```python
func(1, 2, 3)
```

```python
func(2, 3, 4)
```

```python
func(3, 2, 1)
```
