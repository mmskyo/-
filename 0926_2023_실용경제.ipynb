- 그래프 시각화

  $$ y_1 = 2x $$
  $$ y_2 = (1/3)x^2+5 $$
  $$ y_3 = -x^2-5 $$

```python
x = np.arange(-20, 20) # -20 부터 20까지 1간격으로 생성하라

y1 = 2*x
y2 = (1/3)*x**2+5
y3 = -x**2-5

plt.plot(x, y1, x, y2, x, y3)
plt.axis([-30, 30, -30, 30]) #훨씬 더 보기 좋게. 범위를 정함으로서 효과적으로 시각화
plt.show()
```

```python
x = np.arange(-20, 20) # -20 부터 20까지 1간격으로 생성하라

y1 = 2*x
y2 = (1/3)*x**2+5
y3 = -x**2-5

plt.plot(x, y1, 'g--', x, y2, 'r^-', x, y3, 'b*-') #'g--' 그린 컬러로 점선, r^- red로 ^선 ~~
plt.axis([-30, 30, -30, 30]) #훨씬 더 보기 좋게. 범위를 정함으로서 효과적으로 시각화
plt.show()
```

## 데이터 사이의 관련성


   - 변수 : 키, 몸무게, 연간 소득 등과 같이 변하는 양을 표현하는 것
   - 상관관계(correlation) : 두 개의 변수들이 함께 변화하는 관계

   - 상관계수(correlation coefficient)
     - 변수들 사이의 상관관계 정도를 나타내는 수치이다.
       -상관관계가 있는 두 변수가 있을 경우에, 한 값이 증가하면, 다른 값도 증가할 경우 이를 양(+)의 상관관계

       - 음(-)의 상관관계


    - 주의) 상관관계와 인과관계는 차이가 있다.
     - 한 변수가 다른 변수의 직접적인 원인이 되는 것 -> 인과관계
     

## 상관계수를 도출, 시각화
  - numpy의 corrcoef()함수를 사용
  - Pearson correlation 상관계수를 도출 //corr(x, y) = Cov(x, y)/round_x*round*_y

```python
import numpy as np
```

```python
x = np.arange(10)
y1 = x
```

```python
np.corrcoef(x, y1)
```

```python
x = np.arange(10)
y2 = 2 * x
np.corrcoef(x, y2)
```

```python
x = np.arange(10)
y3 = 10 * x
np.corrcoef(x, y3)
```

# 선형임이 중요. 선형적으로 상승한다. 무조건 1이다. y가 뭐가 되었든

- x는 x 그 자체와 동일하다. : 상관계수는 1
- 두 변수는 선형적인 상관관계를 가진다.
  - 모든 성분의 상관계수는 1

```python
x = np.arange(10)
y4 = x**3
np.corrcoef(x, y4). #선형성에 위배됨. 1111이 아님
```

- 난수 : 특정한 규칙이 없는 수

```python
x = np.arange(10)

y5 = np.random.randint(0, 1000, size = 10)
# random integer 정수만 가져와라. 0과 1000사이 10개의 난수를 생성해라.
```

```python
x
```

```python
y5
```

```python
np.corrcoef(x, y5)
```

```python
result = np.corrcoef((x, y5)) #튜플의 형태로 바꾸기 위해 x와 y5에 괄호 삽입
```

```python
y6 = x**3
```

```python
result = np.corrcoef((x, y5, y6))
result #대칭 행렬 symmetirc matrix  x와 x와의 상관계수는 1. y6와는 x의 3승이기에 연관성이 높을 수 박밖에 없다.
```

```python
plt.imshow(result)
plt.colorbar(). #heat map
```
