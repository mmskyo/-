```python
sample = [19, 25, 28, 29, 34, 30, 35, 35, 37, 70, 38]
```

```python
data = sorted(sample)
```

```python
data #아웃라이어 같은 경우 박스 플랏으로 확인. 정렬하고 어쩌구.. raw data를 꼭 확인 필수
```

```python
dataQ1 = np.percentile(data, 25)
```

```python
dataQ1 #29
```

```python
dataQ1, dataQ3 = np.percentile(data, [25, 75])
```

```python
print(dataQ1, dataQ3)
```

```python
IQR = dataQ3 - dataQ1
IQR
```

```python
ul = dataQ3 + IQR*1.5
ll = dataQ1 - IQR*1.5
print(ul, ll)
```
