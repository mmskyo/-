## 논리연산

 - 참 / 거짓
 - 참: 무엇인가 맞다
 - 거짓: 틀리다/ 아니다 표현

 - 참(True), 거짓(False)을 나타내는 연산의 논리 : Boolean

     - and: 둘다 참일 때만 참
     - or: 둘 중 하나만 참이여도 참

     -not: 논리 not 연산
