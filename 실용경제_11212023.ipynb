# Returns

- one-period return

  - one period simple return

  $$R_t = \frac{p_t-p_{t-1}}{p_{t-1}}$$
  =$$\frac{p_t}{p_{t-1}}-\frac{p_{t-1}}{p_{t-1}}$$
  =$$\frac{p_t}{p_{t-1}}-1$$

  - one-period gross return
  $$\frac{p_t}{p{t-1}} = R_t + 1$$

  #### Multiple period return

- holding period가 1단위 기간 이상일 때

**k-period simple return from time $t-k$ to $k$**

$$R_t(k) = \frac{p_t-p_{t-k}}{p_{t-k}}$$

이게 위에 보다 더 제너럴하다. k단위니까



- k-period gross return

$$\frac{p_t}{p_{t-k}} = R_t(k)+1$$


$$1+R_t(k) = (1+R_t)\times (1+R_{t-1})\times \cdots \times (1+R_{t-k+1})$$

$$R_t(k) = (1+R_t)\times (1+R_{t-1})\times \cdots \times (1+R_{t-k+1}) - 1$$

```python
((data.pct_change()+1).cumprod()).plot() #위처럼 각각의 데타에 1씩을 더한 것  cumprod=cummulated product
plt.legend()
plt.title("Gross Returns")
plt.ylabel("cumulative returns")
plt.xlabel("year")
plt.show()
```

```python
df = data["Adj Close"]
```

```python
df.head()
```

```python
df.columns = ["samsung", "msft", "S&P", "kospi"]
```

```python
df.head()
```

```python
df.iloc[-2] #-1하면 미국시장이 아직 안열려서 불가능 오늘 20일
```

```python
df.iloc[0] #integer location showing index 0th value
```

```python
cum_return = (df.iloc[-2]-df.iloc[0])/df.iloc[0]
cum_return * 100 #gross return 구하는 법임
```

```python
df.head()
```

```python
df_daily_ret = df.pct_change()
df_daily_ret
```

```python
df_daily_ret = df_daily_ret[1:]
df_daily_ret #NaN 빼버리기
```

```python
# cumulative daily return = R_t(k)

df_cum_daily_ret = (df_daily_ret + 1).cumprod() - 1 #이거가지고 그래프도 그려라
```

```python
cum_ret_entire_period = df_cum_daily_ret.tail(1)
cum_ret_entire_period #5년간 축적된 수익률
```

```python
df_daily_ret[["samsung", "msft"]].plot()
plt.title("Daily simple returns for samsung elec. and msft")
plt.show()
```

```python
df_cum_daily_ret[["samsung", "msft"]].plot()
plt.title("Cumulative returns for samsung elec. and msft")
plt.show()
```
