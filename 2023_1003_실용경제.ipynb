## Figure, axes에 대하여 학습

```python
fig, ax = plt.subplots(2, 2) #2행과 2열의 그래프를 그리는 공간을 설정

x = np.random.randn(100) #정규분포(randn)를 가지는 데이터(np.random 랜덤의 데이터)
y = np.random.rand(100) #평균0, 표준편차 1의 가우시안 분포
ax[0, 0].scatter(x, y) #0번째 행과 0번째 열에 x와 y로 scatter plot을 그리겠다.

x = np.arange(10) #0과 9사이의 값. 1씩 증가
y = np.random.uniform(1, 10, 10) #uniform distribution 확률이 uni. 균일 분포 값 생성(최소,최대,몇개)
ax[0, 1].bar(x, y) #10분의 1 확률로 10개 생성

x = np.linspace(0, 10, 100)
y = np.cos(x)
ax[1, 0].plot(x, y)

z = np.random.uniform(0, 1, (5, 5)) # 0,1,2,3,4 총 5개로 그래프
ax[1, 1].imshow(z) #2D 이미지 그림 분포

plt.show()
```

## 자료 값의 분포를 나타내는 산점도와 막대 그래프

- 산점도 그래프 (scatter plot)
  - 두 변수의 상관관계를 직교 좌표계(x-y축)의 평면에 점으로 표현하는 그래프

  - matplotlib.pyplot 모듈의 scatter() 사용
  - plt.scatter()


  - plt.scatter(x, y)

```python
import matplotlib.pyplot as plt
import numpy as np

n = 50;
x = np.random.rand(n) #rand 함수는 0과 1사이의 범위에서 난수를 n개 생성(괄호 안 변수만큼)
y = np.random.rand(n)

plt.scatter(x, y, color = "g")
plt.show()
```

- plt.scatter(x, y, colors)

- 색상과 크기를 산점도 그래프에서 지정하기

  - plt.scatter(x, y, area, color) #area = size of the circle

  - scatter() 함수의 s는 마커(동그라미)의 크기를 의미한다.
  - scatter() 함수의 c는 마커의 색상을 지정

```python
n = 50

x = np.random.rand(n)
y = np.random.rand(n)
area = (30*np.random.rand(n))**2 #원의 사이즈 설정
colors = np.random.rand(n) #임의의 숫자를 rgb코드로 변환
plt.scatter(x, y, s = area, c = colors, alpha = 0.5) #alpha = 농도. 겹쳐지는 것을 확인
```

- bar plot

```python
x = np.arange(3)

years = ["2021", "2022", "2023"]
domestic = [6800, 7680, 8010]
foreign = [888, 1047, 1689]

plt.bar(x, domestic)
plt.bar(x, foreign)
plt.xticks(x, years) #x 대신에 years
plt.show()
```

```python
x = np.arange(3)

years = ["2021", "2022", "2023"]
domestic = [6800, 7680, 8010]
foreign = [888, 1047, 1689]

plt.bar(x, domestic, width = 0.25, label = "domestic")
plt.bar(x + 0.3, foreign, width = 0.25, label = "foreign") #따로따로 그리기 위하여 x축의 포지션을 너비보다 넓은 간격으로 더하기
plt.xticks(x, years)
plt.xlabel("years")
plt.ylabel("numbers")
plt.legend()
plt.show()

# width 키워드 인자는 막대 너비를 0.25로 축소
```

```python
x = np.arange(3)

years = ["2021", "2022", "2023"]
domestic = [6800, 7680, 8010]
foreign = [888, 1047, 1689]

plt.barh(x, domestic, height = 0.25, label = "domestic") #bar 뒤에 h(meaning horizontal) 붙이면 가로 그래프
plt.barh(x + 0.3, foreign, height = 0.25, label = "foreign")
plt.legend()
plt.title("numbers")
plt.yticks(x, years) #y가 아니라 x 써야함. 축의 의미가 아닌 우리가 x = np.arange(3)이라고 명명한 거니까
plt.show()
```

## 파이차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [5, 3, 7, 14]
list = ["cyan", "gray", "orange", "red"]

plt.pie(data, colors = list, autopct = '%.2f%%') #auto percent 소수점 둘쨋자리 까지 나오게
```

- pie() 함수의 인자 autopct:
  - 부채꼴 안에 표시될 숫자의 형식을 지정
  - %.2f : 소수점 아래 둘째 자리까지 표시
  - %까지 표시하고 싶을 때 : %.2f%%(%%두개 연속 입력)

```python
data = [5, 3, 7, 14]
list = ["cyan", "gray", "orange", "red"]

plt.pie(data, colors = list, autopct = '%.2f%%', labels = list) #리스트를 그대로 사용하겠다. 반시계방향
plt.show()
```

## 히트맵
