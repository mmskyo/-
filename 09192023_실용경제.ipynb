#함수는 값을 return할 수 있고, 또 안해도 된다.

- return이 없는 경우

```python
def my_func(x, y) :
  print(x, y)
```

```python
my_func(10, 1000)
```

- return이 있는 경우

```python
def my_func3(x, y) :
  z = x + y
  return z
```

```python
my_func3(10, 100)
```

- (x + y) * 2z 라는 함수

```python
def my_func(x, y, z) :
  return (x + y) * 2 * z
```

```python
my_func(2, 3, 4)
```

- return이 있는 경우에 변수에 값을 다시 할당할 수 있음

```python
#ret(return) 리턴값을 다시 사용가능
ret = my_func3(10, 100)
ret * 11
```
