### Data normalization

- Min-Max approach 값이 너무 커서 normalization하는 것임

$$ x_{mm} = \frac{x-x_{min}}{x_{max}-x_{min}}$$

```python
df["length"].head()
```

```python
df["length"] = (df["length"]-df["length"].min())/(df["length"].max()-df["length"].min())
df["length"]
```

```python
#z-score. also normalization scale down
df["length"] = (df["length"]-df["length"].mean())/df["length"].std()
df["length"].head()
```

# Binning

- 연속적 수치형 데이터를 비연속적(discrete)인 범주형 데이터로 변환

```python
len(df["horsepower"].value_counts()) #len = length 59개가 있다.
```

```python
df["horsepower"].unique()
```

```python
df["horsepower"] = df["horsepower"].astype(int, copy = True) #copy = False integer로 다 바뀌는데 실수형만 그대로
df["horsepower"].unique()
```

```python
df["horsepower"].value_counts()
```

```python
plt.hist(df["horsepower"]) #bin 나누기 전에 히스토그램 그려보기
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower")
plt.show()
```

```python
list_horsepower = df["horsepower"].unique().tolist()
```

```python
min(list_horsepower)
```

```python
max(list_horsepower)
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

- pd.cut(X, bins, labels) 메소드를 활용하여 연속형 변수의 여러개 구간별 범주화 시도

```python
group_name = ["low", "medium", "high"]

df["horsepower-binned"] = pd.cut(df["horsepower"], bins, labels = group_name,
                                 include_lowest = True)

df["horsepower-binned"]
```

```python
df["horsepower-binned"].value_counts()
```

```python
plt.bar(group_name, df["horsepower-binned"].value_counts(), color = "purple")
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```
